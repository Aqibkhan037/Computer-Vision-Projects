```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread('/content/index.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.show()
```

```python
edges = cv2.Canny(gray, 100, 200)
plt.imshow(edges, cmap='gray')
plt.title("Edge Map")
plt.show()
```

```python
import numpy as np
lines = cv2.HoughLines(edges, 1, np.pi/180, 150)
for rho,theta in lines[:,0]:
    a,b = np.cos(theta), np.sin(theta)
    x0, y0 = a*rho, b*rho
    x1, y1 = int(x0 + 1000*(-b)), int(y0 + 1000*(a))
    x2, y2 = int(x0 - 1000*(-b)), int(y0 - 1000*(a))
    cv2.line(img, (x1,y1), (x2,y2), (0,0,255), 2)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Hough Lines Detected")
plt.show()
```

```python
import numpy as np

accumulator = np.zeros((200, 180))
y_idxs, x_idxs = np.nonzero(edges)
for i in range(len(x_idxs)):
    x, y = x_idxs[i], y_idxs[i]
    for theta in range(0,180):
        rho = int(x*np.cos(np.deg2rad(theta)) + y*np.sin(np.deg2rad(theta)))
        if 0 <= rho < 200:
            accumulator[rho,theta] += 1

plt.imshow(accumulator, cmap='hot')
plt.title("Accumulator Voting Space (r, θ)")
plt.xlabel("θ")
plt.ylabel("r")
plt.show()

```

```python
lines_p = cv2.HoughLinesP(edges, 1, np.pi/180, 100, minLineLength=50, maxLineGap=10)
for x1,y1,x2,y2 in lines_p[:,0]:
    cv2.line(img, (x1,y1), (x2,y2), (0,255,0), 2)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Probabilistic Hough Lines")
plt.show()
```

```python
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, dp=1, minDist=30,
                           param1=100, param2=30, minRadius=20, maxRadius=100)

circles = np.uint16(np.around(circles))
for x,y,r in circles[0,:]:
    cv2.circle(img, (x,y), r, (255,0,0), 2)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Hough Circles Detected")
plt.show()
```
